# 8. Data transformation using apply() & map() functions

```python
df
```

```python
# apply() function can be applied over the complete Dataframe
# We can apply apply() on more then one column

df1['Salary_per_Age'] = df1.apply(lambda row: row['Salary'] / row['Age'], axis = 1)  # divides the salary by age
df1
```

### Merging two dataframes

```python
# merge() function merge the dataframe based upon common column
# merge() function behaves like the join operation in sql
df1 = pd.DataFrame({'ID' : [1,2,3], 'Hero' : ['Allu arjun', 'Vijay', 'Ram']})
df2 = pd.DataFrame({'ID' : [2,3,4], 'NetWorth' : [300, 250, 100]})
merged_df = pd.merge(df1,df2, on = "ID")
merged_df
```

```python
merged_df = pd.merge(df1,df2, on = "ID", how = 'inner')
merged_df
```

```python
# left merge : take all the data of leftside table, but only common data from the right side table
merged_df = pd.merge(df1,df2, on = "ID", how = 'left')
merged_df
```

```python
# right merge : this will take the commplete data of right side table, but only common data from the left side table
merged_df = pd.merge(df1,df2, on = "ID", how = 'right')
merged_df
```

```python
merged_df = pd.merge(df1,df2, on = "ID", how = 'outer')
merged_df
```

```python
joined_df = df1.join(df2, how = 'left')
joined_df
```

# 10. Time series  related functions

```python
# Generating synthetic time series data
date_rng = pd.date_range(start = '2020-01-01', end = '2025-12-31',  freq = 'h')
date_rng
```

```python
# Selecting 1000 records from above date time data

import numpy as np
np.random.seed(42)  # here we take any seed value 
random_dates = np.random.choice(date_rng, size = 1000, replace = False)
print(len(random_dates))
random_dates
```

```python
# Create a synthetic time series dataset

df = pd.DataFrame({
    'timestamp' : pd.to_datetime(random_dates),
    'order_id' : np.arange(1, 1001),
    'category': np.random.choice(['Biryani', 'Paya', 'Irani Chai', 'Maska Bun'], size = 1000),
    'Quantity': np.random.randint(1,5, size = 1000),
    'price' : np.random.randint(25, 450, size = 1000)
})
df
```

```python
df['Sales'] = df['price'] * df['Quantity']
df
```

```python
df = df.sort_values('timestamp').reset_index(drop = True)
df
```

```python
df['Year'] = df.timestamp.dt.year
```

```python
df['Month'] = df.timestamp.dt.month
```

```python
df['Day'] = df.timestamp.dt.day
```

```python
df['Hour'] = df.timestamp.dt.hour
```

```python
df['Minute'] = df.timestamp.dt.minute
```

```python
df['Second'] = df.timestamp.dt.second
df
```
